### ‚á® Edad

```python
# funcion para juntar las edades en grupos
def categoria_edad(x):
    x = int(x)
    x = abs(x)
    if( 18 < x < 31 ):
        return "19-30"
    if( 30 < x < 41 ):
        return "31-40"
    if( 40 < x < 51 ):
        return "41-50"
    if( 50 < x < 61 ):
        return "51-60"
    if( 60 < x < 71 ):
        return "61-70"
    else:
        return "Mayor que 70"

# aplicandolo a una nueva columna
adultos['categoria_edad'] = adultos['edad'].apply(categoria_edad)
```

```python
# indicando el tamano del marco
fig, ax =  plt.subplots(figsize=(15, 4))

# creando groupby para generar el grafico
educacion_y = adultos.groupby(['educacion', 'ingreso_y']).size().unstack(fill_value=0).reset_index()
educacion_y = educacion_y.rename(columns={0: '<=50K', 1: '>50K'})

# cambiandolo de formato
educacion_y_long = educacion_y.melt(id_vars='educacion', value_vars=['<=50K', '>50K'], var_name='variables').sort_values('value', ascending=False)

# creando el grafico
sns.barplot(data=educacion_y_long, x='educacion', y='value', hue='variables', palette='Blues_r')

# configurando titulos y variables
plt.title('Conteo de ingresos por educacion', fontsize = 18, fontweight = 'bold')
plt.xlabel('Educacion')
plt.ylabel('Ganancias por personas')
plt.xticks(rotation=75)

# A√±adir etiquetas de datos a cada barra
for p in ax.patches:
    ax.annotate(str(p.get_height()),
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='center',
                xytext=(0, 5),
                textcoords='offset points',
                fontsize=8)

plt.show()
```

**An√°lisis de Porcentajes**

En este estudio, analizaremos las probabilidades que tiene cada variable de esta columna para ganar m√°s o menos de 50,000 d√≥lares al a√±o.

```python
# calculando las probabilidades
educacion_y['total_educacion'] = educacion_y['<=50K'] + educacion_y['>50K']
educacion_y['probabilidad_>50K'] = ((educacion_y['>50K'] / educacion_y['total_educacion']) * 100).round(2)
educacion_y['probabilidad_<=50K'] = ((educacion_y['<=50K'] / educacion_y['total_educacion']) * 100).round(2)

#conviertiendo a formato float
educacion_y['probabilidad_>50K'] = educacion_y['probabilidad_>50K'].astype(float)
educacion_y['probabilidad_<=50K'] = educacion_y['probabilidad_<=50K'].astype(float)
```

```python
# Agrupar los datos por la columna "educacion" y calcular la suma de las columnas de porcentaje
grouped_data = educacion_y.groupby("educacion").sum()

# Obtener las categor√≠as y las columnas de porcentaje
categorias = grouped_data.index
columnas = ["probabilidad_>50K", "probabilidad_<=50K"]

# Configurar los datos para cada columna
datos_columnas = [grouped_data[col] for col in columnas]

# Crear el gr√°fico de columnas apiladas
fig, ax =  plt.subplots(figsize=(15, 4))
plt.bar(categorias, datos_columnas[0], label="Probabilidad de ganar m√°s de 50K dentro de la educacion")
plt.bar(categorias, datos_columnas[1], bottom=datos_columnas[0], label="Probabilidad de ganar menos o igual de 50K dentro de la educacion")

# Agregar etiquetas de datos a cada barra
for i in range(len(categorias)):
    x = categorias[i]
    y1 = datos_columnas[0][i]
    y2 = datos_columnas[1][i]
    if columnas[1] == "probabilidad_<=50K":
        plt.text(x, y1/2, f"{y1:.2f}%", ha='center', va='center')
        plt.text(x, y1 + y2/2, f"{y2:.2f}%", ha='center', va='center', color='white')
    else:
        plt.text(x, y1/2, f"{y1:.2f}%", ha='center', va='center')
        plt.text(x, y1 + y2/2, f"{y2:.2f}%", ha='center', va='center')

# Configurar etiquetas y leyendas
plt.xlabel("Educacion")
plt.ylabel("Porcentaje")
plt.title("Probabilidad de ganar mas de 50K de cada educacion", fontsize=18, fontweight='bold')
plt.legend()
plt.xticks(rotation=75)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar el gr√°fico
plt.show()
```

El valor F1, que combina tanto la precisi√≥n como el recall, refleja adecuadamente esta situaci√≥n. Al tener un recall bajo, el F1 tambi√©n se ve afectado negativamente, ya que se calcula como la media arm√≥nica entre precisi√≥n y recall. Por lo tanto, si uno de ellos es bajo, el F1 tambi√©n lo ser√°.
